## <b>Analyzing the Data (EDA)
<b>Exploratory Data Analysis (EDA)</b> is the process of examining a dataset to summarize its main characteristics, often using visual methods or quick commands. The goal is to understand the structure of the data, detect patterns, spot anomalies, and get a feel for what kind of preprocessing or modeling might be needed.

For our project, we’ll perform EDA on the California housing dataset (which we are treating as if it represents Gurgaon data). Here are some key commands we’ll use:

### 4. `df.describe()`
- Provides statistical summaries for numeric columns.

- Shows:

- <b>Count</b>: Total number of non-null entries
- <b>Mean</b>: Average value
- <b>Std</b>: Standard deviation
- <b>Min</b>: The smallest value (0th percentile or 1st quartile in some contexts)
- <b>25%</b>: The 1st quartile (Q1) — 25% of the data is below this value
- <b>50%</b>: The median or 2nd quartile (Q2) — half of the data is below this value
- <b>75%</b>: The 3rd quartile (Q3) — 75% of the data is below this value
- <b>Max</b>: The largest value (often considered the 4th quartile or 100th percentile)

<i>"<b>Percentiles</b> divide the data into 100 equal parts. <b>Quartiles</b> divide the data into 4 equal parts (Q1 = 25th percentile, Q2 = 50th, Q3 = 75th). So:"</i>"

- <b>Min</b> is the 0th percentile
- <b>Max</b> is the 100th percentile

```python
#train and test dataframes 
import numpy as np
```

```python
def shuffle_and_split(data , test_ratio):
    #seed function freeze the training data for reproducibility
    np.random.seed(42)
    # Shake the bucket to mix the blocks
    shuffled_indices = np.random.permutation(len(data)) 
    
    # Decide how many blocks go in the test pile
    test_set_size = int(len(data) * test_ratio)
    
    # Pick the first few for testing
    test_indices = shuffled_indices[:test_set_size]
    
    #Pick the REST of the blocks for training (start from where test ended)
    train_indices = shuffled_indices[test_set_size:]
    
    # Give back the two separate piles
    return data.iloc[train_indices] , data.iloc[test_indices]
```

```python
train , test = shuffle_and_split(df , 0.4)
```
