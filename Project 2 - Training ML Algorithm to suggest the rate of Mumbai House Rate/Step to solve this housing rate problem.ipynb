## <B>Revisiting Steps to solve this problem
### <b>Understanding the Problem
Before we build the model, we need to understand what kind of machine learning problem we are solving.

First, we’ll check if this is a < B>supervised</b> or <b>unsupervised</b> learning task. Since we have historical data with house prices (our target), and we want to predict the price of a house based on input features, this is a <b>supervised learning</b> problem.

Next, we observe that the model is predicting <b>one continuous label</b> — the price of a house — based on several input features. This makes it a <b>univariate regression problem.</b>
